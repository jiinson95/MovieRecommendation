**User-based Collaborative Filtering**
<br>
if a person A likes item 1,2,3 and B likes 2,3,4 then they have similar interets and A should like item 4 and B should like item 1

```python
# read base and test dataset
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings_base = pd.read_csv('ua.base', sep='\t', names=r_cols, encoding='latin-1')
ratings_test = pd.read_csv('ua.test', sep='\t', names=r_cols, encoding='latin-1')
ratings_base.shape, ratings_test.shape
```

```python
# copy the base dataset into ratings_dict
ratings_dict = ratings_base.copy()
```

```python
# make a dictionary of users and all the movies they rated
all_dict = dict()
for usr in ratings_base.user_id:
    one_usr = ratings_dict[ratings_dict.user_id == usr]
    usr_dict = dict(zip(one_usr.movie_id, one_usr.rating))
    all_dict[str(usr)] = usr_dict
```

```python
data = all_dict
# function for calculating similarities
def euclidean_similarity(person1, person2):
    #find common items ranked by both person1 and person2
    common_ranked_items = [itm for itm in data[person1] if itm in data[person2]]
    rankings = [(data[person1][itm], data[person2][itm]) for itm in common_ranked_items]
    distance = [pow(rank[0] - rank[1], 2) for rank in rankings]

    return 1 / (1 + sum(distance)) # the range is from 0 to 1
```

```python
# create a recommendation function that will recoomend movies to users

def recommend(person, bound, similarity=euclidean_similarity):
    scores = [(similarity(person, other), other) 
              for other in data if other != person] # calculate the similarity score between the person and others
    
    scores.sort() 
    scores.reverse() #descending order
    scores = scores[0:bound] # get the people with the top scores
    # scores is the lists of euclidean score and the other user id
    
    #print (scores)

    recomms = {}

    for sim, other in scores:
        ranked = data[other] # data of the other users which include the movieID and ranking

        for itm in ranked: 
            if itm not in data[person]: #if item is not in the data of user 1 then weight = sim score * 
                weight = sim * ranked[itm] #multiplying the similarity score with the ranking of that movie id
                
                if itm in recomms:
                    s, weights = recomms[itm]
                    recomms[itm] = (s + sim, weights + [weight])
                    # 
                else:
                    recomms[itm] = (sim, [weight])
               
    for r in recomms: 
        sim, item = recomms[r]
        recomms[r] = sum(item) / sim #predicted ranking for each movie = actual ranking/similarity
        # the reason is because we want to predict how likely each user will like another movie
        # ex: user 1: ABCD, user 2: ABFG
        # how likely will user 1 like movie F & G

    # return recomms
    return sorted(recomms.items(), key=lambda x:x[1], reverse=True)[0:bound]
```

```python
# recommend 10 movies to user_id "1"
results = recommend("1", 10, euclidean_similarity)
print(results)
```

```python
# find actual movie title
train = list()
i = 0
for movie in items['movie id']:
    for order in results:
        i += 1
        if order[0] == movie:
            print (items['movie title'][movie-1])
            train.append(items['movie title'][movie-1])

```
